```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv('shinkansen.csv', thousands=',')
```

```python
df
```

```python
# updating the column value/data
df['Status'] = df['Year'].replace({df['Year']: df['Year'] - 1964})
```

```python
#change the size of the graph
sns.set(rc = {'figure.figsize':(60,40)})
ax = sns.barplot(x="Distance from Tokyo st", y="Year", data=df)
ax.axes.set_title("Title",fontsize=50)
ax.set_xlabel("Distance from Tokyo st",fontsize=30)
ax.set_ylabel("Year",fontsize=20)
```

```python
#change the size of the graph
sns.set(rc = {'figure.figsize':(60,40)})
ax = sns.barplot(x="Year", y="Distance from Tokyo st", data=df)
ax.axes.set_title("Title",fontsize=50)
ax.set_xlabel("Year",fontsize=30)
ax.set_ylabel("Distance from Tokyo st",fontsize=20)
```
